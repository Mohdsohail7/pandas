```python
fruit_protein = {
    "Avocado": 2.0,
    "Guava": 2.6,
    "Orange": 0.9,
    "Banana": 1.1,
    "Apples": 0.3,
    "Kiwi": 1.1,
    "Pomegranate": 1.7,
    "Mango": 0.8,
    "Cherries": 1.0
}
```

```python
s2 = pd.Series(fruit_protein, name = "Protein")
s2
```

```python
# Conditional selection.

s2>1
```

```python
s2[s2>1]
```

```python
# Logical Opertaors: and, or, not
[(s2 > 0.5) & (s2 < 2)]
```

```python
s2[(s2 > 0.5) & (s2 < 2)]
```

```python
s2[(s2 > 0.5) | (s2 <= 2)]
```

```python
# not operator.
s2[~(s2>1)]
```

```python
df
```

```python
# dealing with duplicate data
```

```python
df_dup = df[df.duplicated()] # without pass parameter give the last value as a duplicate (work from top to bottom)
df_dup
```

```python
df_dup = df[df.duplicated(keep= "first")] # (work from top to bottom)
df_dup
```

```python
df_dup = df[df.duplicated(keep= "last")] # mark 1st value as a duplicate (work from bottom to top)
df_dup
```

```python
df = df.drop_duplicates() # this remove second record of same
df
```

```python
# Invalid values
# Lambda
df["Promoted Salary"] = df["Promoted Salary"].apply(lambda x: x/10 if x > 650000 else x)
df
```

```python
# apply and lamda examples
def multiplying_age(x):
  return x*2

df["Age"] = df["Age"].apply(multiplying_age)
df
```

```python
df["Age"] = df["Age"].apply(lambda x: x/2)
df
```

```python
# Joins and Merges
department_info = {
    "Department": ["HR", "IT", "Finance"],
    "Location": ["New York", "San Francisco", "Chicago"],
    "Manager": ["Laura", "Steve", "Nina"]
}
df2 = pd.DataFrame(department_info)
df2
```

```python
# Concat
pd.concat([df, df2]) # this give the result row wise
```

```python
pd.concat([df, df2], axis= 1) # this give result column wise
```

```python
pd.merge(df, df2, on= "Department")
```
